```python
def get_moves_from_video(video_path):
    if not os.path.exists(video_path):
        print(f"Video not found: {video_path}")
        return

    cap = cv2.VideoCapture(video_path)
    
    previous_fen = None
    move_list = []
    
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(frame_rate / 2) # Process 2 frames per second (every 0.5s)
    
    frame_count = 0
    
    print("Processing video...")
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break # End of video
            
        frame_count += 1
        
        # Only process at your desired interval (e.g., every 15 frames if 30fps)
        if frame_count % frame_interval == 0:
            
            # --- This is your core task ---
            current_fen = find_fen_from_frame(frame)
            # -------------------------------

            if previous_fen is None:
                # This is the first frame we've processed
                print("Got starting FEN...")
                previous_fen = current_fen
            
            # THE MOVE DETECTION LOGIC
            if current_fen != previous_fen:
                # A move was made!
                timestamp_ms = cap.get(cv2.CAP_PROP_POS_MSEC)
                timestamp_sec = timestamp_ms / 1000.0
                
                print(f"✅ MOVE DETECTED! at {timestamp_sec:.2f}s")
                print(f"   Old FEN: {previous_fen}")
                print(f"   New FEN: {current_fen}")
                
                # You would also add your "find what move" logic here
                move_list.append((current_fen, timestamp_sec))
                
                previous_fen = current_fen
                
                # Wait a bit so the move is "finished" before checking again
                time.sleep(1) 

    cap.release()
    print("Video processing complete.")
    return move_list
```
